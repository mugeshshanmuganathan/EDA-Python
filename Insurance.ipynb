```python
sns.boxplot(y=df['age'])
```

```python
for i in df.columns:
    if df[i].dtypes =='int64' or  df[i].dtypes =='float64':
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.show()
```

```python
for i in df.columns:
    if df[i].dtypes =='int64' or  df[i].dtypes =='float64':
        Q1=df[i].quantile(0.25)
        Q3=df[i].quantile(0.75)
        IQR= Q3-Q1
        Upper_fen= Q3+1.5*IQR
        lower_fen= Q1-1.5*IQR
        df = df[ (df[i] >= lower_fen) & (df[i] <= Upper_fen)]
        
```

```python
for i in df.columns:
    if df[i].dtypes =='int64' or  df[i].dtypes =='float64':
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.show()
```
